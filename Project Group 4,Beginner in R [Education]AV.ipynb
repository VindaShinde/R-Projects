# Section 9: Multiple Linear Regression

I will be performing Multiple linear regressions having several independant variables together with dependant variable and interpret their results with simple linear regression models

```r
# lm() to run a linear regression and assigned to multi_model1 variable
multi_model1 <- lm(Chance.of.Admit ~ ., data = dataClean)
# Plot the residuals in function of your fitted observations
plot(x= multi_model1$fitted.values, y = multi_model1$residuals)
# Make a Q-Q plot of your residual quantiles
qqnorm(multi_model1$residuals)
#prints the summary of model
summary(multi_model1)
```

**From the analysis above we can observe that, University rating and SOP have P-values greater than 0.05 so these independent variables do not have direct linear dependancy on Chance of admit. So we will create new Model with only relevant variables where P-value is lower than 0.05**


```r
# lm() to run a linear regression and assigned to multi_model1 variable
new_multi_model1 <- lm(Chance.of.Admit ~ GRE.Score + TOEFL.Score + LOR + CGPA + Research,dataClean)

# Plot the residuals in function of our fitted observations
plot(y = new_multi_model1$residuals, x= new_multi_model1$fitted.values)

# creating a Q-Q plot of our residual quantiles
qqnorm(new_multi_model1$residuals)

# prints the summary
summary(new_multi_model1)
```

The new model shows the P-value very much closer to 0 (<<0.05) which indicates we have a stronger relation between Y variable **Chance of Admit** to the x variables **LOR, TOEFL Score, GRE, CGPA and Research.**

**Comparing Multiple linear regression result with simple linear regression result, we can interpret that,Multiple linear regression model shows us the better predictability with highest Adjusted R squared value of 0.8188 also with all the p values << 0.05 in the new_multi_model1, where as the simple linear regression models predicts highest adjusted R squared value of 0.778, predicted with CGPA against chance of admission.**

In overall, it is observed that , SOP rating and University rating has the least significance to the chance of admit, whereas LOR, TOEFL Score, GRE, CGPA and Research together have the highest level of significance on Chance of Admit.
